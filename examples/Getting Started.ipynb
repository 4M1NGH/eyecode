Import the included datasets, area of interest (AOI), and plotting functions.

```python
from eyecode import data, aoi, plot
```

Load fixation data and the screen image from a single trial (experiment 2, trial 11). Display a scatter plot of fixations and saccades.

```python
fixes = data.hansen_2012.all_fixations(2, 11)
img = data.hansen_2012.trial_screen(2, 11)
plot.fix_circles(fixes, img)
```

Get areas of interest for this trial and overlay the "line" AOIs on the screen image.

```python
aois = data.hansen_2012.areas_of_interest(2, 11)
plot.draw_rectangles(aois[aois.kind == "line"], img)
```

Load fixations on individual code lines for this trial. Render a highlighted image of the code and plot relative fixation durations for each line.

```python
lines = data.hansen_2012.line_fixations(2, 11)
code = data.hansen_2012.program_code("rectangle", "class")
code_img = plot.highlight_code(code, filename="code.py",
                               font_name="Droid Sans Mono",
                               line_numbers=True)

plot.line_code_image(lines, code_img, 21)
```

Plot a "barebones" timeline of fixations on each line over the entire trial.

```python
ax = plot.fix_timeline(lines, 21, barebones=True)
```

Generate a scanpath from line fixations, and compute/plot transition probabilities between lines.

```python
sp = aoi.scanpath_from_fixations(fixes, "line", repeats=False)
trans_mat = aoi.transition_matrix(sp)
name_map = { i: "line {0}".format(i+1) for i in range(21) }
ax = plot.aoi_transitions(trans_mat, name_map, show_probs=False)
```

Import functions for fixation/saccade metrics.

```python
from eyecode import metrics
```

Combine line AOIs into a single rectangle.

```python
new_aois = aoi.combine_aois(aois, "line", new_kind="all-code")
code_bbox = new_aois[new_aois.kind == "all-code"]
plot.draw_rectangles(code_bbox, img)
```

Compute the spatial density over a 10x10 grid of the code. This is done with a rolling window 4 seconds in size in increments of 1 second over the entire trial.

```python
import pandas

# Spatial density is computed over the entire code box
def code_density(fs):
    return metrics.spatial_density(fs, grid_bbox=code_bbox, num_cols=10, num_rows=10)

# Combine 3 rolling metrics into a single DataFrame
results = pandas.DataFrame({
    "density": metrics.rolling_window(fixes, code_density, 4000, 1000).value,
    "num_fixes": metrics.rolling_window(fixes, metrics.fixation_count, 4000, 1000).value,
    "duration": metrics.rolling_window(fixes, metrics.avg_fixation_duration, 4000, 1000).value
})

# Plot metrics from DataFrame
ax = plot.rolling_metrics(results, columns=["density", "num_fixes", "duration"],
                          names=["Spatial Density", "Fixation Count", "Avg. Fix Duration"],
                          figsize=(15, 5))
ax.set_title("Rolling Window of Metrics (4+1s)")
ax.grid()
```
