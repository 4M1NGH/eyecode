# Hit Testing

```python
import pandas
from eyecode import aoi, plot
```

## Finding Rectangles

We'll start by loading an image with text (code or otherwise).

```python
lorem = Image.open("img/lorem.png")
lorem
```

Because the text in this image does not have a monospace font, finding the words and lines requires more effort. The `aoi.find_rectangles` method will extract words and lines from a black and white image.

```python
lorem_aois = aoi.find_rectangles(lorem)
lorem_aois
```

By default, only whitespace-separated words (sub-line) and line AOIs are identified.

```python
lorem_aois.kind.unique()
```

Use `plot.draw_rectangles` to view the AOI rectangles.

```python
print "Just the lines"
line_aois = lorem_aois[lorem_aois.kind == "line"]
plot.draw_rectangles(line_aois, lorem)
```

```python
print "Just the words"
word_aois = lorem_aois[lorem_aois.kind == "sub-line"]
plot.draw_rectangles(word_aois, lorem)
```

## Hit Testing Fixations

We'll now create a fake set of fixations by assuming a participant reads all of the words in order (and fixates precisely on the center of each for 200 ms).

```python
simple_scanpath = pandas.Series(word_aois.name)
simple_scanpath
```

```python
simple_fixes = aoi.fixations_from_scanpath(simple_scanpath, word_aois, aoi_kinds="sub-line")
simple_fixes
```

Viewing the fixations is easy with `plot.fix_circles`.

```python
plot.fix_circles(simple_fixes, lorem)
```

Now we have some fixations to hit-test against the AOI rectangles. Passing in all of our AOIs will simultaneously hit-test the words and lines, adding `aoi_sub-line` and `aoi_line` columns to the fixations DataFrame.

```python
aoi_fixes = aoi.hit_test(simple_fixes, lorem_aois, hit_fun="point")
aoi_fixes
```

We can easily see the the amount of time spent on each line with `plot.aoi_barplot`, which will be proportional to the length of the line in this case.

```python
plot.aoi_barplot(aoi_fixes, aoi_kinds="line")
```

A timeline of fixations on each line can be plotted by first calling `aoi.to_line_fixations` to get a DataFrame with an additional `line` column whose value is the integer line number of the fixation.

Then, `plot.fix_timeline` will display all fixations by line number.

```python
line_fixes = aoi.to_line_fixations(aoi_fixes)
plot.fix_timeline(line_fixes, 6)
```
