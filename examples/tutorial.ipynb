```python
from eyecode import data, aoi, plot, metrics
```

Let's start by loading fixations from the very first trial (experiment 1, trial 1).

```python
fix_groups = list(data.hansen_2012.all_fixations()\
    .groupby(["exp_id", "trial_id"]))

(exp_id, trial_id), trial_fixes = fix_groups[0]
```

We can see that there are 103 fixations in this trial.

```python
print "Experiment", exp_id, ", Trial", trial_id
print trial_fixes
```

Let's visualize the fixations (and saccades) by loading a screen shot of the trial and plotting fixation circles on it. The circles' radii are proportional to their duration.

```python
trial_image = data.hansen_2012.trial_screen(trial_id)
plot.fix_circles(trial_fixes, trial_image)
```

Areas of interest (AOIs) have already been identified, and are available with the data. AOIs are grouped by kind into layers.

```python
all_aois = data.hansen_2012.areas_of_interest()
trial_aois = all_aois[(all_aois.exp_id == exp_id) &
                      (all_aois.trial_id == trial_id)]

print "AOI kinds:", trial_aois.kind.unique()
```

AOIs are just named rectangles with an upper-left origin (x, y) and size (width, height).

```python
trial_aois
```

Let's visualize the interface AOIs by drawing their corresponding rectangles on the trial screen shot.

```python
interface_aois = trial_aois[trial_aois.kind == "interface"]
plot.draw_rectangles(interface_aois, trial_image)
```

The red rectangle is called the "code box", while the purple and blue rectangles are the "output box" and "continue button" respectively.

We can calculate some AOI-based metrics easily, such as the number of fixations per AOI.

```python
metrics.fixes_per_aoi(trial_fixes)
```

It's easy to here that most of the fixations on interface AOIs were within the code box. Each AOI kind refers to a different layer, so the same fixations are counted in, for example, the "block" and "line" layers.

We can just as easily determine how much time was spent per AOI (i.e., total fixation duration).

```python
metrics.fixes_ms_per_aoi(trial_fixes)
```

The "line" AOIs are of special interest. Let's first see what they look like.

```python
line_aois = trial_aois[trial_aois.kind == "line"]
plot.draw_rectangles(line_aois, trial_image)
```

An important question we can answer with these AOIs is: did the programmer read the program from top to bottom? To answer this, let's first get the code from this trial.

```python
base, version = trial_fixes[["base", "version"]].values[0]
trial_code = "".join(data.hansen_2012.program_code(base, version))
print "Code for", base, version, "program"
print trial_code
```

Next, we can create an image with this code, highlighted using Pygments.

```python
code_image = plot.highlight_code(trial_code, line_numbers=True)
code_image
```

Next, we'll use the `line_fixations()` data method to get a filtered version of this trial's fixations. These are fixations that:

1. Occurred on a line
2. Have a special `line` property with the 1-based line number

```python
line_fixes = data.hansen_2012.line_fixations()
trial_line_fixes = line_fixes[(line_fixes.exp_id == exp_id) &
                              (line_fixes.trial_id == trial_id)]
trial_line_fixes
```

```python
print "Fixated lines"
print trial_line_fixes.line
```

Finally, we'll use the `line_code_image` plot function to combine our highlighted code with a bar plot indicating the time of the first fixation on each line.

```python
options = dict(num_lines=9, bar_height=0.7,
               bar_mult=1.035, image_padding=15)

plot.line_code_image(trial_line_fixes, code_image,
                     method="first", **options)
```

With this plot, we can readily see several things. First, although the programmer started on line 7, they read through mostly from top to bottom. Second, they did not fixate line 8 until later in the trial.

Using a timeline plot, we can confirm this and note that the programmer visited the output box twice before fixating on line 8 (and one more time after).

```python
output_fixes = trial_fixes[trial_fixes.aoi_interface == "output box"]
plot.fix_timeline(trial_line_fixes, 9, output_fixes=output_fixes)
```

Just as easily, we can find out which lines were fixated on for the most time.

```python
plot.line_code_image(trial_line_fixes, code_image,
                     method="time", **options)
```

An overview of time spent on all AOIs at once can be seen using the `aoi_barplot` function.

```python
ax = plot.aoi_barplot(trial_fixes, method="time", figsize=(10, 5))
ax.set_title("Total Fixation Time by AOI")
```

Using the `aoi` module, we can create a *scanpath* from the fixated lines. This is essentially a timestamped series of fixated AOIs, sorted by fixation time.

```python
line_sp = aoi.scanpath_from_fixations(trial_fixes, aoi_names="line", repeats=False)
line_sp
```

We can compute a transition matrix between pairs of fixated lines using the `transition_matrix` function. Note that repeated fixations on the same line were excluded from the scanpath (using the `repeats=False` option).

```python
line_tm = aoi.transition_matrix(line_sp)
line_tm
```

Plotting the transition matrix, it's easy to see that line 1 -> line 2 and line 7 -> line 6 are the most common transitions in this trial.

```python
name_map = { i : "line {0}".format(i+1) for i in range(9) }
plot.aoi_transitions(line_tm, name_map=name_map, figsize=(10, 8))
```
