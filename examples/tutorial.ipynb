### Creating AOIs from Code

You can create AOIs from your own code easily if it was displayed with a monospace font. For example, let's say we have the following Pythong code:

```python
code = """a = 3
b = 5
print a + b"""
```

Using the `aoi.code_to_aois` function, we can obtain AOI rectangles for every *token* and *line* in the source code. The filename parameter is required to properly guess the code's programming language (more on that in a moment).

```python
aois = aoi.code_to_aois(code, filename="test.py")
aois
```

The AOIs above assume a particular font size, padding, and line spacing. Looking at the help for the `code_to_aois` function reveals the details.

```python
help(aoi.code_to_aois)
```

By default, we assume a font size of 18 pixels, a line offset of 5 pixels, and a padding of 2x5 pixels (2 horizontal, 5 vertical).

We can see what this code would look like when rendered with the default monospace font using `plot.draw_code`.

```python
lines = code.split("\n")
img = plot.draw_code(lines, font_size=18, line_offset=5)
img
```

The *token* AOI kind represents each lexically-separable chunk of the code. Here are the token AOIs rendered on top of the image above:

```python
aoi_img = plot.draw_rectangles(aois[aois.kind == "token"], img)
aoi_img.save("output/aois-token.png")
aoi_img
```

We also get *line* AOIs out of the `code_to_aois` function.

```python
plot.draw_rectangles(aois[aois.kind == "line"], img)
```

Building on top of these basic AOIs is easy. For example, we could add a new "block" AOI by combining the 3 line AOIs

```python
names = aois[aois.kind == "line"].name
aois = aoi.combine_aois(aois, "line", names, "block", "block 1")
plot.draw_rectangles(aois[aois.kind == "block"], img)
```

Creating AOIs from code works on any [language that pygments supports](http://pygments.org/docs/lexers/)! Here's a simple Java example:

```python
code = """package hello;

class Hello {
    public static void main(String[] args) {
        System.out.println("Hello World");
    }
}"""

aois = aoi.code_to_aois(code, filename="hello.java")
lines = code.split("\n")
img = plot.draw_code(lines, font_size=18, line_offset=5)
aoi_img = plot.draw_rectangles(aois[aois.kind == "token"], img)
aoi_img.save("output/aois-java.png")
aoi_img
```

The AOIs will be named according to the Pygments token kind. An attempt at categorizing each token is done using `eyecode.aoi.get_token_category`.

```python
aois
```

## Plotting

An important question we can answer with these AOIs is: did the programmer read the program from top to bottom? To answer this, let's first get the code from this trial.

```python
base, version = trial_fixes[["base", "version"]].values[0]
trial_code = "".join(data.hansen_2012.program_code(base, version))
print "Code for", base, version, "program"
print trial_code
```

Next, we can create an image with this code, highlighted using Pygments.

```python
code_image = plot.highlight_code(trial_code, filename="sample.py",
                                 line_numbers=True, font_name="FreeMono")
code_image
```

Next, we'll use the `line_fixations()` data method to get a filtered version of this trial's fixations. These are fixations that:

1. Occurred on a line
2. Have a special `line` property with the 1-based line number

```python
line_fixes = data.hansen_2012.line_fixations()
trial_line_fixes = line_fixes[(line_fixes.exp_id == exp_id) &
                              (line_fixes.trial_id == trial_id)]
trial_line_fixes
```

```python
print "Fixated lines"
print trial_line_fixes.line
```

Finally, we'll use the `line_code_image` plot function to combine our highlighted code with a bar plot indicating the time of the first fixation on each line.

```python
options = dict(num_lines=9, bar_height=0.7,
               bar_mult=1.035, image_padding=15)

img = plot.line_code_image(trial_line_fixes, code_image,
                           method="first", **options)
img.save("output/line-code-first.png")
img
```

With this plot, we can readily see several things. First, although the programmer started on line 7, they read through mostly from top to bottom. Second, they did not fixate line 8 until later in the trial.

Using a timeline plot, we can confirm this and note that the programmer visited the output box twice before fixating on line 8 (and one more time after).

```python
output_fixes = trial_fixes[trial_fixes.aoi_interface == "output box"]
ax = plot.fix_timeline(trial_line_fixes, 9, output_fixes=output_fixes)
ax.figure.savefig("output/fix-timeline.png")
```

Just as easily, we can find out which lines were fixated on for the most time.

```python
img = plot.line_code_image(trial_line_fixes, code_image,
                           method="time", **options)
img.save("output/line-code-time.png")
img
```

An overview of time spent on all AOIs at once can be seen using the `aoi_barplot` function.

```python
aoi_kinds = ["interface", "line", "syntax", "block"]
ax = plot.aoi_barplot(trial_fixes, method="time", aoi_kinds=aoi_kinds, figsize=(10, 5))
ax.set_title("Total Fixation Time by AOI")
```

## Scanpaths

Using the `aoi` module, we can create a *scanpath* from the fixated lines. This is essentially a timestamped series of fixated AOIs, sorted by fixation time.

```python
line_sp = aoi.scanpath_from_fixations(trial_fixes, aoi_names="line", repeats=False)
line_sp
```

We can compute a transition matrix between pairs of fixated lines using the `transition_matrix` function. Note that repeated fixations on the same line were excluded from the scanpath (using the `repeats=False` option).

```python
line_tm = aoi.transition_matrix(line_sp)
line_tm
```

Plotting the transition matrix, it's easy to see that line 1 -> line 2 and line 7 -> line 6 are the most common transitions in this trial.

```python
name_map = { i : "line {0}".format(i+1) for i in range(9) }
ax = plot.aoi_transitions(line_tm, name_map=name_map, figsize=(10, 8), show_probs=False)
ax.figure.savefig("output/aoi-transitions.png")
```

## Rolling Metrics

```python
rw1 = metrics.rolling_window(trial_saccs, metrics.avg_saccade_length, 1000, 500).dropna()
rw2 = metrics.rolling_window(trial_fixes, metrics.avg_fixation_duration, 1000, 500).dropna()
rw = rw1.join(rw2, lsuffix="_sacc", rsuffix="_fix")
rw
```

```python
ax = plot.rolling_metrics(rw, ["value_sacc", "value_fix"],
                          names=["Avg. Saccade Length", "Avg. Fixation Duration"],
                          figsize=(10, 5))
ax.grid()
ax.set_title("Rolling Metrics for Experiment {0} Trial {1}".format(exp_id, trial_id))
```
